# 8 Sneakers from FlipKart

```python
driver = webdriver.Chrome('chromedriver.exe')
driver
```

```python
url = 'https://www.flipkart.com'
driver.get(url)
```

```python
search = driver.find_element_by_xpath("//input[@type='text']")
```

```python
search.send_keys('sneakers')
```

```python
search_click = driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
```

```python
#fetch all the urls from difference pages
urls = []
for u in driver.find_elements_by_xpath("//nav[@class='yFHi8N']/a"):
    urls.append(u.get_attribute('href'))
```

```python
# user urls collected to scrap the data according to requirement
Brand = []
Product_Description = []
Price = []
Discount = []
for i in urls:
    driver.get(i)
    
    for b in driver.find_elements_by_xpath("//div[@class='_2WkVRV']"):
        Brand.append(b.text)
    for p in driver.find_elements_by_xpath("//a[@class='IRpwTa']"):
        Product_Description.append(p.text)
    for m in driver.find_elements_by_xpath("//div[@class='_30jeq3']"):
        Price.append(m.text)
    for d in driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']/span"):
        Discount.append(d.text)
```

```python
#create a dataframe to list down the scrapped details
Sneakers = pd.DataFrame({})
Sneakers['Brand']=Brand[:100]
Sneakers['Product Description']=Product_Description[:100]
Sneakers['Price']=Price[:100]
Sneakers['Discount']=Discount[:100]
```

```python
Sneakers
```
